## 함수
임베딩 모델 생성 함수 모음
```
embedding_texts() : 
  toeknized_text → make_tokenizer_tf() → make_embedding_weights()
  
  return (tokenized_tf, embedding_weights)
```

````python
import tensorflow as tf
import numpy as np
````

### 문자 표현 → 정수 표현 모델 (tokenizer_tf)

````python
def make_tokenizer_tf(tokenized_texts) :
  tokenizer_tf = tf.keras.preprocessing.text.Tokenizer()
  tokenizer_tf.fit_on_texts(tokenized_texts)

  return tokenizer_tf
````

### 정수 표현 → 벡터 표현 모델 (embedding_weights)

````python
def make_embedding_weights(fasttext, tokenizer_tf) :
  # tf.keras.layers.Embedding 에 넣을 수 있는 형태로 바꾼다
  embedding_weights = np.random.random((len(tokenizer_tf.word_index) + 1, fasttext.vector_size))
  
  for word,i in tokenizer_tf.word_index.items():
      try:
        embedding_weights[i] = fasttext.wv[word]    # [[벡터 표현], [벡터 표현], ...] <- 이런식으로 만듬
      
      # fasttext 사전에 word가 없을 경우
      except:
        print(word)
        pass

  return embedding_weights
````

### 문자 표현 → 정수 표현 → 벡터 표현 (embedding_texts)

````python
def make_embedding_models(tokenized_texts, fasttext) :
  tokenizer_tf = make_tokenizer_tf(tokenized_texts)

  embedding_weights = make_embedding_weights(fasttext, tokenizer_tf)  

  return (tokenizer_tf, embedding_weights)
````
